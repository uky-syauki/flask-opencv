```python
import cv2 as cv
import os
import time

def setDariRekam(User):
    camera = cv.VideoCapture('vidio/recorded.mp4')
    face_detec = cv.CascadeClassifier('cascade/face_detect.xml')
    count = 0
    ulang = 0

    while True:
        ulang += 1
        ret, img = camera.read()
        
        # Menghentikan jika video telah selesai
        if not ret:
            break

        # img = cv.flip(img, -1)
        gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
        wajah = face_detec.detectMultiScale(gray, 1.1, 5)

        for (x, y, w, h) in wajah:
            count += 1
            cv.imwrite(f"dataset/{User['nama']}/{str(count)}.jpg", gray[y:y+h, x:x+w])

        cv.imshow("image", img)
        
        if ulang >= 15:
            break

        # Jeda sebelum mengambil gambar berikutnya
        time.sleep(1)  # Jeda selama 1 detik (sesuaikan dengan kebutuhan Anda)

        # Press 'q' untuk keluar
        if cv.waitKey(1) & 0xFF == ord('q'):
            break

    camera.release()
    cv.destroyAllWindows()

setDariRekam({'id': 1, 'nama': 'uki'})
```
